```python
replace with mean and median generally we replace with median because mean can be affected by outliar so that can skew the distrbution of  and data median is robust to outliars
```

```python
genreally beginer replace all value with avg of all reveiw of all category but different category carries different rting so we replace perticular category median with null 
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# now change the columns name 
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels=['young adult','young','middle age','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
print(df[['age','age_group']].head())
```

```python
mapping={'Weekly':7,
         'Fortnightly':14,
         'Monthly':30,
         'Quarterly':90,
         'Bi-Weekly':14,
         'Annually':365,
         'Every 3 Months':90}
df['purchase_frequency_day']=df['frequency_of_purchases'].map(mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_day']]
```

```python
df[['discount_applied','promo_code_used']]
```

```python
df=df.drop('promo_code_used',axis=1)
```
